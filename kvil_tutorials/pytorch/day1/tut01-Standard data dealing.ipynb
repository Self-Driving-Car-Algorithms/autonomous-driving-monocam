```python
from PIL import Image
import os

if not os.path.exists('tut01-results'):
    os.makedirs('tut01-results')

#Loading an Image
img = Image.open('data/tut01/iitkgp.jpg')
img.show() #display the image
print "Image size: ", str(img.size)
```

```python
#converting image to grayscale
img_gray = img.convert('L')
img_gray.show()
```

```python
#cropping the image
box = (250, 10, 450, 250)
img_crop = img.crop(box)
img_crop.show()
```

```python
#resizing the image
img_resz = img.resize((200, 200))
img.show()
img_resz.show()
```

```python
#rotating the image
img_rot = img.rotate(45)
img_resz.show()
img_rot.show()
```

```python
%matplotlib inline
from pylab import *

figure(1)
hist(array(img_gray).flatten(), 256)
show()
```

```python
import matplotlib.pyplot as plt

img_grayarr = array(img.convert('L'), 'f')
img_grayinv = 255 - img_grayarr

plt.subplot(2, 2, 1)
imshow(img_grayarr, 'gray')

plt.subplot(2, 2, 2)
imshow(img_grayinv, 'gray')

plt.subplot(2, 2, 4)
imshow(img, 'gray')

img_clamp = (100.0/255) * img_grayarr + 100
figure()
imshow(img_clamp)
plt.colorbar()
```

```python
import cv2 as cv
img_cv = cv.imread('data/tut01/iitkgp.jpg')

plt.subplot(1, 2, 1)
imshow(img_cv)

plt.subplot(1, 2, 2)
imshow(img)
```
